```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

```python
img = cv2.imread('C:/Users/kalpe/AI_ML Project/Face Mask Detection/img2.jpg')
```

```python
plt.imshow(img)
```

```python
while True:
    cv2.imshow('result',img)
    # 2 -> millisecond of time 
    if cv2.waitKey(2) == 13: # 13 -> ASCII of Enter , 27 -> Esc 
        break
    
cv2.destroyAllWindows()
```

```python
haar_data = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
haar_data.detectMultiScale(img)
#cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
```

```python
# for drawing rectangle 

while True:
    faces = haar_data.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w, y+h),(255,0,255),2)
    cv2.imshow('result',img)
    # 2 -> millisecond of time 
    if cv2.waitKey(2) == 13: # 13 -> ASCII of Enter , 27 -> Esc 
        break
cv2.destroyAllWindows()
```
